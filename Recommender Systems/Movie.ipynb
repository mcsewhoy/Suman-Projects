```python
artists = pd.read_csv('artists.csv')
user_artists = pd.read_csv('user_artists.csv')
user_taggedartists = pd.read_csv('user_taggedartists.csv')
tags = pd.read_csv('tags.csv')
#Merged_table = pd.read_csv('Merged_table.csv')
```

```python
user_taggedartists.drop(['day','month','year'],axis=1,inplace=True)
```

```python
user_taggedartists = user_taggedartists.pivot_table(index="artistID",columns="tagID", values="userID", aggfunc='count', fill_value=0)
```

```python
user_taggedartists.head()
```

```python
user_taggedartists = user_taggedartists.add_prefix("tagID_")
```

```python
user_taggedartists.reset_index(inplace=True)
```

```python
user_taggedartists.head()
```

```python
whole_df = artists.copy(deep=True)
```

```python
whole_df = whole_df.rename(mapper={'id':'artistID'},axis=1)
```

```python
whole_df.drop(['url','pictureURL'],axis=1, inplace=True)
```

```python
whole_df.head()
```

```python
whole_df = whole_df.merge(user_taggedartists, on='artistID')
```

```python
whole_df.head()
```

```python
user_artists = user_artists.pivot_table(values="weight",index="artistID",columns="userID",fill_value=0)
```

```python
user_artists = user_artists.add_prefix("userID_")
```

```python
user_artists.reset_index(inplace=True)
```
